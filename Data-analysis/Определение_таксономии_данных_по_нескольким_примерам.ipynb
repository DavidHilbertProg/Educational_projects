```python
import pandas as pd
```

```python
# Сохраняем данные файлов
data1 = pd.read_csv('data1.csv')
data2 = pd.read_csv('data2.csv')
data3 = pd.read_csv('data3.csv')
```

```python
# С помощью ".head()" можно посмотреть первые строки в данных
data1.head()
```

```python
# С помощью ".info()" смотрим какие типы данных в файле
data1.info()
```

```python
cortege = (data1, data2, data3)
for i in cortege:
    print()
    i.info()
```

```python
# Другой вариант
```

```python
# Определение таксономии данных
taxonomy = {
    'data1': data1.dtypes,
    'data2': data2.dtypes,
    'data3': data3.dtypes
}
```

```python
# Вывод результатов таксономии данных
for dataset, dtypes in taxonomy.items():
    print("Таксономия данных для", dataset)
    print(dtypes)
    print()
```
