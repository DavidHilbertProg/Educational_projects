```python
import pandas as pd
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Download file
data = pd.read_csv('price.csv')
```

```python
# Предварительный анализ данных
# Проверка наличия пропущенных значений
missing_values = data.isnull().sum()
print("Пропущенные значения:")
print(missing_values)
```

Среднее значение

```python
mean = np.mean(data.Value)
np.mean(data.Value)
```

Среднеквадратическое (стандартное) отклонение

```python
std = np.std(data.Value)
np.std(data.Value)
```

Размах:

```python
range = np.max(data.Value) - np.min(data.Value)
np.max(data.Value) - np.min(data.Value)
```

Квартили

```python
q1 = np.quantile(data.Value, q=0.25)
q3 = np.quantile(data.Value, q=0.75)
q1, q3
```

```python
# Оценка статистических характеристик данных
statistics = data.describe()
print("Статистические характеристики:")
print(statistics)
```
